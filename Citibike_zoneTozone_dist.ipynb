```python
import pandas as pd

citi8 = pd.read_csv('201808-citibike-tripdata.csv')
citi9 = pd.read_csv('201809-citibike-tripdata.csv')
citi10 = pd.read_csv('201810-citibike-tripdata.csv')
citi11 = pd.read_csv('201811-citibike-tripdata.csv')
#citi8.head()
```

```python
citi_total=pd.concat([ citi8,citi9, citi10, citi11])
citi_total.head()
```

```python
citi_total_station=citi_total[['start station id','end station id','bikeid']]
tripCountPerStation=citi_total_station.groupby(['start station id','end station id']).count()
tripCountPerStation.head()
```

```python
tripCountPerStation.to_excel("citibike_start_end_station_zipcode.xlsx")
```
